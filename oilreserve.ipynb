```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
df=pd.read_csv('/home/msois/Downloads/oil_reserves.csv')
df.head(8)
```

```python
# Which region has max oil reserves from 2014 to 2016?
df.max()[["Region", "Country/ Region","2014", "2015", "2016"]]
```

```python
#With suitable visualization show the oil reserves of India from 2010 to 2016.
# df[df["Country/ Region"]=="India"]["2010", "2011" ,"2012" ,"2013","2014", "2015", "2016"].plot(kind='bar')
# plt.show()
df.groupby("Country/ Region"=='India')['2010', "2011" ,"2012" ,"2013","2014", "2015", "2016"].plot(kind='bar')
plt.show()
```

```python
# List the countries having maximum oil reserves in each region.
df.max()[["Region", "Country/ Region","2010", "2011" ,"2012" ,"2013","2014", "2015", "2016"]].head()
# df.nlargest(5, ["Region", "Country/ Region","2010", "2011" ,"2012" ,"2013","2014", "2015", "2016"]).head()
```

```python
# Draw suitable visualization which shows the share of oil reserves by each region.
df.groupby("Region")['2010', "2011" ,"2012" ,"2013","2014", "2015", "2016"].plot(kind='bar')
plt.show()
```

```python
# Find the world’s top 5 countries having maximum oil reserves in the 2015. Show their % of share
# in total oil reserve for the same year.
df.nlargest(5, ['2015'])
```
